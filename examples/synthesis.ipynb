# EEG recoding to fMRI volume

In this post I will go over the procedure to project an EEG instance to an fMRI. This work is the core of my PhD thesis and all the findings that consequently followed. The motivation of synthesizing an fMRI volume from EEG rests on:
- An EEG recording setup is cheaper than an MRI laboratory setup;
- The fMRI recording session in itself costs hundreds of dollars, whereas an EEG recording does not;
- The main hypothesis of this work is to develop an informed fMRI view that allows better diagnosis than the EEG view alone, consequently having the potential of being applied in health care systems and reducing costs.

## What is the model?

In the [EEG to fMRI paper](https://arxiv.org/abs/2203.03481), the methodology to do this projection is presented.

This model processes two inputs:
- EEG representation $\vec{x} \in \mathbb{R}^{C \times F \times T}$;
- fMRI volume representation $\vec{y} \in \mathbb{R}^{M_1 \times M_2 \times M_3}$.

## What is an EEG?

The EEG has a drift in relation to the associated fMRI volume, since it takes into consideration $10\times \mbox{TR}$ seconds in total (for the [NODDI](https://osf.io/94c5t/) dataset this corresponds to $10\times 2.160=21.6$ seconds). In addition, the EEG has $C$ channels and $F$ frequency coefficients.

## What is an fMRI?

On the other hand, we have a **single** fMRI volume associated with the respective EEG. This fMRI is described by three dimensions, corresponding to the 3-dimensional axis, where $M_1=64$, $M_2=64$ and $M_3=30$.

With these parameters, we can finally build the model and setup the optimizer, loss, training set and test set:

````python
from eeg_to_fmri.models.synthesizers import na_specification_eeg
from eeg_to_fmri.models.fmri_ae import na_specification_fmri

with tf.device('/CPU:0'):
    model = EEG_to_fMRI(latent_dimension, eeg_train.shape[1:], na_specification_eeg, n_channels,
                        weight_decay=weight_decay, skip_connections=True, batch_norm=True, fourier_features=True,
                        random_fourier=True, topographical_attention=True, conditional_attention_style=True,
                        conditional_attention_style_prior=False, local=True, seed=None,
                        fmri_args = (latent_dimension, fmri_train.shape[1:], kernel_size, stride_size, n_channels,
                        max_pool, batch_norm, weight_decay, skip_connections,
                        n_stacks, True, False, outfilter, dropout, None, False, na_specification_fmri))
    model.build(eeg_train.shape, fmri_train.shape)
    optimizer = tf.keras.optimizers.Adam(learning_rate)
    loss_fn = losses.mae_cosine
    train_set = tf.data.Dataset.from_tensor_slices((eeg_train, fmri_train)).batch(batch_size)
    test_set= tf.data.Dataset.from_tensor_slices((eeg_test, fmri_test)).batch(1)
````

Note that the loss used minimizes the objective of approximating the EEG with the fMRI, along with the latent representations of each other, meaning:

$$\mathcal{L}(\vec{x}, \vec{y}) = ||\vec{y}-\hat{y}||_1^1  + 1-\frac{\vec{z}_x \cdot \vec{z}_y}{||\vec{z}_x||_2^2\cdot ||\vec{z}_y||_2^2}$$

This loss approximates the output (predicted fMRI) with the ground truth (fMRI volume) using the L1 distance and approximates the latent representations with a pattern based (cosine) distance.

Finally, comes the fun part! The model is trained with the objective of producing an fMRI volume similar to the one paired with the given EEG:

````python
import tensorflow as tf
from eeg_to_fmri.learning import train, losses
from eeg_to_fmri.models.synthesizers import EEG_to_fMRI

# 1. æ¸…ç©ºèˆŠçš„è¨˜æ†¶é«”ï¼Œé¿å…æ··äº‚
tf.keras.backend.clear_session()

# ç¢ºä¿åƒæ•¸æ­£ç¢º (å†æ¬¡å®šç¾©ä»¥é˜²è¬ä¸€)
latent_dimension = (10, 1, 1) # è¨˜å¾—è¦ç”¨ Tuple
learning_rate = 0.0002
batch_size = 8

print("ğŸš€ æ­£åœ¨å°‡æ¨¡å‹ç§»è‡³ GPU é‡å»º...")

# 2. é—œéµä¿®æ”¹ï¼šæ”¹ç”¨ GPU (æˆ–è€…è®“ TF è‡ªå‹•æ±ºå®š)
# ç§»é™¤ '/CPU:0'ï¼Œæ”¹ç”¨ '/GPU:0'
with tf.device('/GPU:0'):
    model = EEG_to_fMRI(
        latent_dimension,
        eeg_train.shape[1:],
        na_specification_eeg,
        n_channels,
        weight_decay=weight_decay,
        skip_connections=True,
        batch_norm=True,
        fourier_features=True,
        random_fourier=True,
        topographical_attention=True,
        conditional_attention_style=True,
        conditional_attention_style_prior=False,
        local=True,
        seed=None,
        fmri_args=(
            latent_dimension,
            fmri_train.shape[1:],
            kernel_size,
            stride_size,
            n_channels,
            max_pool,
            batch_norm,
            weight_decay,
            skip_connections,
            n_stacks,
            True,
            False,
            outfilter,
            dropout,
            None,
            False,
            na_specification_fmri
        )
    )

    # å»ºç«‹æ¨¡å‹æ¬Šé‡
    model.build((None,) + eeg_train.shape[1:], (None,) + fmri_train.shape[1:])

    # é‡æ–°å®šç¾©å„ªåŒ–å™¨ (é€™ä¹Ÿè¦åœ¨ GPU ä¸Š)
    optimizer = tf.keras.optimizers.Adam(learning_rate)
    loss_fn = losses.mae_cosine

print("âœ… æ¨¡å‹å·²åœ¨ GPU å°±ç·’ï¼")

# 3. é‡æ–°å»ºç«‹ Dataset (ç¢ºä¿æ‰¹æ¬¡å¤§å°æ­£ç¢º)
# é€™è£¡ä¸éœ€è¦æ”¹ï¼Œtf.data æœƒè‡ªå‹•è™•ç†æ¬ç§»
train_set = tf.data.Dataset.from_tensor_slices((eeg_train, fmri_train)).batch(batch_size)

# 4. é–‹å§‹è¨“ç·´
print("ğŸš€ é–‹å§‹è¨“ç·´ (Epochs = 10)...")
train.train(
    train_set,
    model,
    optimizer,
    loss_fn,
    epochs=10,
    u_architecture=True,
    val_set=None,
    verbose=True,
    verbose_batch=True
)
````

````python
import tensorflow as tf
import numpy as np

# ==================== 1. ä¿®æ­£å¾Œçš„ MC é æ¸¬å‡½æ•¸ ====================

def predict_MC(model, x, y, T=10):
    """
    Monte Carlo Prediction:
    é æ¸¬ T æ¬¡ä¸¦å–å¹³å‡ã€‚
    ä¿®æ­£ï¼šå‚³å…¥ y (fmri) ä»¥æ»¿è¶³æ¨¡å‹åœ¨ training=True æ™‚çš„è¼¸å…¥è¦æ±‚ã€‚
    """
    y_hat_sum = None

    # é æ¸¬ T æ¬¡
    for _ in range(T):
        # é€™è£¡å¿…é ˆåŒæ™‚å‚³å…¥ x (EEG) å’Œ y (fMRI)
        # å³ä½¿æ˜¯é æ¸¬ï¼Œç‚ºäº†æ»¿è¶³ tf.function çš„ç°½ç« ï¼Œæˆ‘å€‘éœ€è¦é¤µçµ¦å®ƒæ ¼å¼æ­£ç¢ºçš„è³‡æ–™
        pred = model(x, y, training=True)

        # è™•ç†è¼¸å‡ºå¯èƒ½æ˜¯ list çš„æƒ…æ³ (é€šå¸¸ [0] æ˜¯æˆ‘å€‘è¦çš„é æ¸¬åœ–)
        if isinstance(pred, list):
            pred = pred[0]

        if y_hat_sum is None:
            y_hat_sum = pred
        else:
            y_hat_sum += pred

    return y_hat_sum / T

# ==================== 2. é–‹å§‹è©•ä¼° ====================

print("ğŸš€ é–‹å§‹é«˜éšè©•ä¼° (ä½¿ç”¨ Monte Carlo æ¡æ¨£, T=10)...")

total_ssim = []
total_rmse = []

# éæ­·æ¸¬è©¦é›†
for step, (eeg_batch, fmri_true_batch) in enumerate(test_set):

    # ã€é—œéµä¿®æ­£ã€‘ï¼šåŒæ™‚å‚³å…¥ eeg_batch å’Œ fmri_true_batch
    try:
        fmri_pred_batch = predict_MC(model, eeg_batch, fmri_true_batch, T=10)
    except Exception as e:
        # è¬ä¸€é‚„æ˜¯å ±éŒ¯ï¼Œå˜—è©¦ç”¨ list åŒ…èµ·ä¾†çš„å¯«æ³• model([x, y])
        print(f"âš ï¸ æ¨™æº–å‘¼å«å¤±æ•—ï¼Œå˜—è©¦ List è¼¸å…¥æ¨¡å¼... ({e})")
        # å‚™ç”¨æ–¹æ¡ˆï¼šæœ‰äº› Keras æ¨¡å‹å–œæ­¡ list è¼¸å…¥
        def predict_MC_list(model, x, y, T=10):
            y_hat_sum = None
            for _ in range(T):
                # å˜—è©¦å‚³å…¥ list
                pred = model([x, y], training=True)
                if isinstance(pred, list): pred = pred[0]
                if y_hat_sum is None: y_hat_sum = pred
                else: y_hat_sum += pred
            return y_hat_sum / T

        fmri_pred_batch = predict_MC_list(model, eeg_batch, fmri_true_batch, T=10)

    # 2. è½‰æ›æ ¼å¼ (Tensor -> Numpy) æº–å‚™è¨ˆç®—
    # ç¢ºä¿ç¶­åº¦ä¸€è‡´
    y_true = fmri_true_batch
    y_pred = fmri_pred_batch

    # 3. è¨ˆç®— RMSE (å‡æ–¹æ ¹èª¤å·®)
    # axis=(1,2,3,4) ä»£è¡¨å° (H, W, Depth, Channel) åšå¹³å‡
    batch_rmse = tf.sqrt(tf.reduce_mean(tf.square(y_true - y_pred), axis=(1,2,3,4)))
    total_rmse.extend(batch_rmse.numpy())

    # 4. è¨ˆç®— SSIM (çµæ§‹ç›¸ä¼¼æ€§)
    # é‡å° 3D å½±åƒï¼Œæˆ‘å€‘é€å±¤è¨ˆç®— SSIM ç„¶å¾Œå–å¹³å‡
    batch_ssim = 0
    depth = y_true.shape[3] # æ·±åº¦é€šå¸¸æ˜¯ 30

    for d in range(depth):
        # å–å‡ºç¬¬ d å±¤åˆ‡ç‰‡: (Batch, 64, 64, 1)
        slice_true = y_true[:, :, :, d, :]
        slice_pred = y_pred[:, :, :, d, :]

        # ç¢ºä¿æ•¸å€¼ç¯„åœæ­£ç¢º (SSIM éœ€è¦çŸ¥é“æœ€å¤§å€¼)
        # ç‚ºäº†é¿å…è¨ˆç®—éŒ¯èª¤ï¼Œé€™è£¡å‹•æ…‹æŠ“å–æœ€å¤§å€¼
        max_val = tf.reduce_max([tf.reduce_max(slice_true), tf.reduce_max(slice_pred)])

        # è¨ˆç®—é€™ä¸€å±¤çš„ SSIM
        batch_ssim += tf.image.ssim(slice_true, slice_pred, max_val=max_val)

    # å¹³å‡æ‰€æœ‰å±¤çš„ SSIM
    batch_ssim /= depth
    total_ssim.extend(batch_ssim.numpy())

    if step % 5 == 0:
        print(f"   å·²è™•ç† Batch {step}...")

# ==================== 3. é¡¯ç¤ºçµæœ ====================
avg_rmse = np.mean(total_rmse)
avg_ssim = np.mean(total_ssim)

print("\n" + "="*30)
print("ğŸ“Š é€²éšè©•ä¼°å ±å‘Š (Advanced Metrics)")
print("="*30)
print(f"ğŸ”¹ RMSE (å‡æ–¹æ ¹èª¤å·®): {avg_rmse:.4f}  (è¶Šä½è¶Šå¥½)")
print(f"ğŸ”¹ SSIM (çµæ§‹ç›¸ä¼¼æ€§): {avg_ssim:.4f}  (è¶Šé«˜è¶Šå¥½ï¼Œæ»¿åˆ† 1.0)")
print("="*30)
````

````python
# ============================================================
# Cell 1: Clone å°ˆæ¡ˆ (ä¿æŒä¸è®Šï¼Œä½†åŠ å…¥æª¢æŸ¥)
# ============================================================
import os

if not os.path.exists('/content/mip_final_project'):
    print("ğŸ“¥ æ­£åœ¨ä¸‹è¼‰å°ˆæ¡ˆ...")
    !git clone https://github.com/xc90529/mip_final_project.git
    print("âœ… å°ˆæ¡ˆä¸‹è¼‰å®Œæˆ")
else:
    print("âœ… å°ˆæ¡ˆå·²å­˜åœ¨ï¼Œè·³éä¸‹è¼‰")

# ============================================================
# Cell 2: Clone è³‡æ–™é›† (ä¿æŒä¸è®Šï¼Œä½†åŠ å…¥æª¢æŸ¥)
# ============================================================
if not os.path.exists('/content/ds002336'):
    print("ğŸ“¥ æ­£åœ¨ä¸‹è¼‰è³‡æ–™é›†...")
    !git clone https://github.com/OpenNeuroDatasets/ds002336.git
    print("âœ… è³‡æ–™é›†ä¸‹è¼‰å®Œæˆ")
else:
    print("âœ… è³‡æ–™é›†å·²å­˜åœ¨ï¼Œè·³éä¸‹è¼‰")

# ============================================================
# Cell 3: å®‰è£å¥—ä»¶ (æ•´åˆå„ªåŒ–ç‰ˆ)
# ============================================================
# ä¸€æ¬¡æ€§å®‰è£æ‰€æœ‰éœ€è¦çš„å¥—ä»¶ï¼Œé¿å…é‡è¤‡å®‰è£
print("ğŸ“¦ é–‹å§‹å®‰è£ç›¸å®¹å¥—ä»¶...")

# ä½¿ç”¨ -q æ¸›å°‘è¼¸å‡ºï¼ŒåŠ å¿«å®‰è£é€Ÿåº¦
!pip install -q matplotlib==3.7.5
!pip install -q nilearn==0.10.0  # æ”¹ç”¨è¼ƒæ–°ä¸”ç©©å®šçš„ç‰ˆæœ¬
!pip install -q scipy==1.11.4
!pip install -q mne
!pip install -q awscli  # æå‰å®‰è£ä¸‹è¼‰å·¥å…·

print("âœ… å¥—ä»¶å®‰è£å®Œæˆï¼")

# ============================================================
# Cell 4: ç’°å¢ƒè¨­å®š (å„ªåŒ–ç‰ˆ)
# ============================================================
import sys
import os
import warnings
warnings.filterwarnings('ignore')  # æ¸›å°‘è­¦å‘Šè¨Šæ¯

# è·¯å¾‘è¨­å®š
project_path = '/content/mip_final_project'
data_path = '/content/ds002336'

# åŠ å…¥æœå°‹è·¯å¾‘
if project_path not in sys.path:
    sys.path.append(project_path)
if f"{project_path}/src" not in sys.path:
    sys.path.append(f"{project_path}/src")

# åˆ‡æ›å·¥ä½œç›®éŒ„
%cd {project_path}

# è¨­å®šç’°å¢ƒè®Šæ•¸
os.environ['EEG_FMRI_DATASETS'] = '/content'
os.environ['EEG_FMRI'] = project_path
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'  # æ¸›å°‘ TensorFlow è­¦å‘Š
os.environ['TF_USE_LEGACY_KERAS'] = '1'   # ä½¿ç”¨èˆŠç‰ˆ Keras

print(f"âœ… å·¥ä½œç›®éŒ„: {os.getcwd()}")
print(f"âœ… è³‡æ–™é›†è·¯å¾‘: {data_path}")

# ============================================================
# Cell 5: è¼‰å…¥ TensorFlow å’Œå°ˆæ¡ˆæ¨¡çµ„ (æ”¹é€²ç‰ˆ)
# ============================================================
import tensorflow as tf
import numpy as np

# æª¢æŸ¥ GPU
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    try:
        # è¨­å®šå‹•æ…‹è¨˜æ†¶é«”å¢é•·ï¼ˆæ›´å¥½çš„æ–¹å¼ï¼‰
        for gpu in gpus:
            tf.config.experimental.set_memory_growth(gpu, True)
        print(f"âœ… æ‰¾åˆ° {len(gpus)} å€‹ GPUï¼Œå·²è¨­å®šå‹•æ…‹è¨˜æ†¶é«”å¢é•·")
    except RuntimeError as e:
        print(f"âš ï¸ GPU è¨­å®šè­¦å‘Š: {e}")
else:
    print("âš ï¸ æœªæ‰¾åˆ° GPUï¼Œå°‡ä½¿ç”¨ CPU")

# è¼‰å…¥å°ˆæ¡ˆæ¨¡çµ„
import eeg_to_fmri
from eeg_to_fmri.utils import tf_config
from eeg_to_fmri.data import preprocess_data, eeg_utils, data_utils
from eeg_to_fmri.models.synthesizers import EEG_to_fMRI
from eeg_to_fmri.learning import train, losses
from eeg_to_fmri import metrics
from eeg_to_fmri.utils import viz_utils

# è¨­å®šéš¨æ©Ÿç¨®å­
tf_config.set_seed(seed=42)  # æ”¹ç”¨æ›´å¸¸è¦‹çš„ç¨®å­

print("âœ… æ‰€æœ‰æ¨¡çµ„è¼‰å…¥æˆåŠŸï¼")

# ============================================================
# Cell 6: ä¸‹è¼‰ fMRI è³‡æ–™ (å„ªåŒ–ç‰ˆ - åŠ å…¥é€²åº¦é¡¯ç¤º)
# ============================================================
import os
from tqdm import tqdm  # é€²åº¦æ¢

data_root = '/content/ds002336'
subjects = [f"sub-xp{i}" for i in range(101, 111)]
task_name = "task-eegfmriNF_bold.nii.gz"

print(f"ğŸ§  é–‹å§‹ä¸‹è¼‰ {len(subjects)} ä½å—è©¦è€…çš„ fMRI è³‡æ–™...")

for sub in tqdm(subjects, desc="ä¸‹è¼‰é€²åº¦"):
    local_dir = os.path.join(data_root, sub, 'func')
    os.makedirs(local_dir, exist_ok=True)

    s3_path = f"s3://openneuro.org/ds002336/{sub}/func/{sub}_{task_name}"
    local_path = os.path.join(local_dir, f"{sub}_{task_name}")

    # æª¢æŸ¥æ˜¯å¦å·²å­˜åœ¨
    if os.path.exists(local_path) and os.path.getsize(local_path) > 1024*1024:
        continue  # è·³éå·²ä¸‹è¼‰çš„æª”æ¡ˆ

    cmd = f"aws s3 cp {s3_path} {local_path} --no-sign-request --quiet"
    exit_code = os.system(cmd)

    if exit_code != 0:
        print(f"âŒ {sub} ä¸‹è¼‰å¤±æ•—")

print("âœ… fMRI è³‡æ–™ä¸‹è¼‰å®Œæˆï¼")

# ============================================================
# Cell 7: ä¸‹è¼‰ EEG è³‡æ–™ (å„ªåŒ–ç‰ˆ)
# ============================================================
print(f"ğŸ§  é–‹å§‹ä¸‹è¼‰ {len(subjects)} ä½å—è©¦è€…çš„ EEG è³‡æ–™...")

for sub in tqdm(subjects, desc="ä¸‹è¼‰é€²åº¦"):
    local_dir = os.path.join(data_root, sub, 'eeg')
    os.makedirs(local_dir, exist_ok=True)

    # æª¢æŸ¥é—œéµæª”æ¡ˆæ˜¯å¦å·²å­˜åœ¨
    vhdr_file = os.path.join(local_dir, f"{sub}_task-eegfmriNF_eeg.vhdr")
    if os.path.exists(vhdr_file):
        continue

    s3_path = f"s3://openneuro.org/ds002336/{sub}/eeg/"
    cmd = f"aws s3 sync {s3_path} {local_dir} --no-sign-request --quiet"
    os.system(cmd)

print("âœ… EEG è³‡æ–™ä¸‹è¼‰å®Œæˆï¼")

# ============================================================
# Cell 8: è³‡æ–™é è™•ç† (é—œéµæ”¹é€²)
# ============================================================
dataset = "ds002336"
n_individuals = 10  # ä½¿ç”¨å…¨éƒ¨ 10 ä½å—è©¦è€…
interval_eeg = 10

print("ğŸš€ é–‹å§‹è¼‰å…¥ä¸¦è™•ç†è³‡æ–™...")
print(f"ğŸ“Š ä½¿ç”¨å—è©¦è€…æ•¸é‡: {n_individuals}")

# ä½¿ç”¨ CPU é€²è¡Œé è™•ç†ï¼ˆè¼ƒç©©å®šï¼‰
with tf.device('/CPU:0'):
    train_data, test_data = preprocess_data.dataset(
        dataset,
        n_individuals=n_individuals,
        interval_eeg=interval_eeg,
        ind_volume_fit=False,
        standardize_fmri=True,  # æ¨™æº–åŒ– fMRI
        iqr=True,  # æ”¹ç‚º Trueï¼šä½¿ç”¨å››åˆ†ä½è·å»é™¤ç•°å¸¸å€¼
        verbose=True
    )

    eeg_train, fmri_train = train_data
    eeg_test, fmri_test = test_data

print("\nâœ… è³‡æ–™è™•ç†å®Œæˆï¼")
print(f"ğŸ“ EEG Train: {eeg_train.shape}")
print(f"ğŸ“ fMRI Train: {fmri_train.shape}")
print(f"ğŸ“ EEG Test: {eeg_test.shape}")
print(f"ğŸ“ fMRI Test: {fmri_test.shape}")

# ============================================================
# Cell 9: å„²å­˜é è™•ç†è³‡æ–™ (ä¿æŒä¸è®Š)
# ============================================================
processed_data_dir = 'processed_data'
os.makedirs(processed_data_dir, exist_ok=True)

print("ğŸ’¾ æ­£åœ¨å„²å­˜é è™•ç†è³‡æ–™...")
np.save(os.path.join(processed_data_dir, 'eeg_train.npy'), eeg_train)
np.save(os.path.join(processed_data_dir, 'fmri_train.npy'), fmri_train)
np.save(os.path.join(processed_data_dir, 'eeg_test.npy'), eeg_test)
np.save(os.path.join(processed_data_dir, 'fmri_test.npy'), fmri_test)

print("âœ… è³‡æ–™å„²å­˜æˆåŠŸï¼")

# ============================================================
# Cell 10: è¼‰å…¥æ¨¡å‹åƒæ•¸ (æ”¹é€²ç‰ˆ - åŠ å…¥åƒæ•¸å„ªåŒ–)
# ============================================================
from eeg_to_fmri.models.synthesizers import parameters

# è¼‰å…¥é è¨­åƒæ•¸
learning_rate, weight_decay, kernel_size, stride_size, batch_size, \
latent_dimension, n_channels, max_pool, batch_norm, skip_connections, \
dropout, n_stacks, outfilter, local = parameters

# ğŸ”§ é—œéµæ”¹é€²ï¼šèª¿æ•´è¶…åƒæ•¸ä»¥æå‡æ•ˆæœ
learning_rate = 0.0001  # é™ä½å­¸ç¿’ç‡ï¼ˆåŸæœ¬å¯èƒ½å¤ªé«˜ï¼‰
batch_size = 16  # å¢åŠ æ‰¹æ¬¡å¤§å°ï¼ˆæ›´ç©©å®šçš„æ¢¯åº¦ï¼‰
dropout = 0.3  # å¢åŠ  dropoutï¼ˆé˜²æ­¢éæ“¬åˆï¼‰

print("ğŸ“‹ æ¨¡å‹è¶…åƒæ•¸ï¼š")
print(f"   Learning Rate: {learning_rate}")
print(f"   Batch Size: {batch_size}")
print(f"   Dropout: {dropout}")
print(f"   Latent Dimension: {latent_dimension}")

# ============================================================
# Cell 11: å»ºç«‹æ¨¡å‹ (æ”¹é€²ç‰ˆ)
# ============================================================
from eeg_to_fmri.models.synthesizers import na_specification_eeg
from eeg_to_fmri.models.fmri_ae import na_specification_fmri

# æ¸…ç©ºèˆŠæ¨¡å‹
tf.keras.backend.clear_session()

print("ğŸ—ï¸ æ­£åœ¨å»ºç«‹æ¨¡å‹...")

with tf.device('/GPU:0'):
    model = EEG_to_fMRI(
        latent_dimension,
        eeg_train.shape[1:],
        na_specification_eeg,
        n_channels,
        weight_decay=weight_decay,
        skip_connections=True,
        batch_norm=True,
        fourier_features=True,
        random_fourier=True,
        topographical_attention=True,
        conditional_attention_style=True,
        conditional_attention_style_prior=False,
        local=True,
        seed=42,  # å›ºå®šç¨®å­
        fmri_args=(
            latent_dimension,
            fmri_train.shape[1:],
            kernel_size,
            stride_size,
            n_channels,
            max_pool,
            batch_norm,
            weight_decay,
            skip_connections,
            n_stacks,
            True,
            False,
            outfilter,
            dropout,
            None,
            False,
            na_specification_fmri
        )
    )

    model.build((None,) + eeg_train.shape[1:], (None,) + fmri_train.shape[1:])

    # ğŸ”§ æ”¹é€²ï¼šä½¿ç”¨å­¸ç¿’ç‡è¡°æ¸›
    lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
        initial_learning_rate=learning_rate,
        decay_steps=1000,
        decay_rate=0.96
    )
    optimizer = tf.keras.optimizers.Adam(lr_schedule)

    # ğŸ”§ æ”¹é€²ï¼šä½¿ç”¨æ··åˆæå¤±å‡½æ•¸
    loss_fn = losses.mae_cosine

print("âœ… æ¨¡å‹å»ºç«‹å®Œæˆï¼")

# å»ºç«‹ Dataset
train_set = tf.data.Dataset.from_tensor_slices((eeg_train, fmri_train))\
    .shuffle(buffer_size=1000)\
    .batch(batch_size)\
    .prefetch(tf.data.AUTOTUNE)

test_set = tf.data.Dataset.from_tensor_slices((eeg_test, fmri_test))\
    .batch(1)

print(f"âœ… Dataset æº–å‚™å®Œæˆ (Batch Size: {batch_size})")

# ============================================================
# Cell 12: è¨“ç·´æ¨¡å‹ (é—œéµæ”¹é€²)
# ============================================================
import time

# ğŸ”§ é—œéµæ”¹é€²ï¼šå¢åŠ è¨“ç·´è¼ªæ•¸
EPOCHS = 50  # å¾ 10 å¢åŠ åˆ° 50

print(f"ğŸš€ é–‹å§‹è¨“ç·´ (Epochs: {EPOCHS})")
print("=" * 50)

start_time = time.time()

# è¨“ç·´
history = train.train(
    train_set,
    model,
    optimizer,
    loss_fn,
    epochs=EPOCHS,
    u_architecture=True,
    val_set=None,  # å¯ä»¥æ”¹ç‚º test_set é€²è¡Œé©—è­‰
    verbose=True,
    verbose_batch=True
)

training_time = time.time() - start_time

print("=" * 50)
print(f"âœ… è¨“ç·´å®Œæˆï¼è€—æ™‚: {training_time/60:.2f} åˆ†é˜")

# å„²å­˜æ¨¡å‹
model.save_weights('model_weights/eeg_to_fmri_improved.h5')
print("ğŸ’¾ æ¨¡å‹æ¬Šé‡å·²å„²å­˜")

# ============================================================
# Cell 13: è¦–è¦ºåŒ–é æ¸¬çµæœ (æ”¹é€²ç‰ˆ)
# ============================================================
import matplotlib.pyplot as plt

print("ğŸ¨ æ­£åœ¨ç”Ÿæˆé æ¸¬è¦–è¦ºåŒ–...")

# é æ¸¬å¤šå€‹æ¨£æœ¬
n_samples = 3  # é¡¯ç¤º 3 å€‹æ¸¬è©¦æ¨£æœ¬

for idx, (eeg, fmri) in enumerate(test_set.take(n_samples)):
    prediction = model(eeg, fmri)[0]

    # ç¹ªè£½ 3D åˆ‡ç‰‡åœ–
    fig = viz_utils.plot_3D_representation_projected_slices(
        prediction.numpy()[0],
        threshold=0.35  # ç•¥å¾®é™ä½é–¾å€¼
    )

    plt.title(f'æ¸¬è©¦æ¨£æœ¬ #{idx+1} - é æ¸¬çµæœ')
    plt.show()

    # ä¹Ÿç¹ªè£½çœŸå¯¦çš„ fMRI
    fig_true = viz_utils.plot_3D_representation_projected_slices(
        fmri.numpy()[0],
        threshold=0.35
    )
    plt.title(f'æ¸¬è©¦æ¨£æœ¬ #{idx+1} - çœŸå¯¦ fMRI')
    plt.show()

# ============================================================
# Cell 14: é€²éšè©•ä¼° (æ”¹é€²ç‰ˆ)
# ============================================================
import tensorflow as tf
import numpy as np
from tqdm import tqdm

def predict_MC(model, x, y, T=20):  # å¢åŠ æ¡æ¨£æ¬¡æ•¸
    """Monte Carlo é æ¸¬"""
    y_hat_sum = None
    for _ in range(T):
        pred = model(x, y, training=True)
        if isinstance(pred, list):
            pred = pred[0]
        if y_hat_sum is None:
            y_hat_sum = pred
        else:
            y_hat_sum += pred
    return y_hat_sum / T

print("ğŸš€ é–‹å§‹é€²éšè©•ä¼°...")

total_ssim = []
total_rmse = []
total_pcc = []  # æ–°å¢ï¼šPearson ç›¸é—œä¿‚æ•¸

for eeg_batch, fmri_true_batch in tqdm(test_set, desc="è©•ä¼°é€²åº¦"):
    # MC é æ¸¬
    fmri_pred_batch = predict_MC(model, eeg_batch, fmri_true_batch, T=20)

    # RMSE
    batch_rmse = tf.sqrt(tf.reduce_mean(
        tf.square(fmri_true_batch - fmri_pred_batch),
        axis=(1,2,3,4)
    ))
    total_rmse.extend(batch_rmse.numpy())

    # SSIM (é€å±¤è¨ˆç®—)
    depth = fmri_true_batch.shape[3]
    batch_ssim = 0
    for d in range(depth):
        slice_true = fmri_true_batch[:, :, :, d, :]
        slice_pred = fmri_pred_batch[:, :, :, d, :]
        max_val = tf.reduce_max([
            tf.reduce_max(slice_true),
            tf.reduce_max(slice_pred)
        ])
        batch_ssim += tf.image.ssim(slice_true, slice_pred, max_val=max_val)
    batch_ssim /= depth
    total_ssim.extend(batch_ssim.numpy())

    # ğŸ”§ æ–°å¢ï¼šPearson ç›¸é—œä¿‚æ•¸
    true_flat = tf.reshape(fmri_true_batch, [-1])
    pred_flat = tf.reshape(fmri_pred_batch, [-1])
    pcc = tfp.stats.correlation(true_flat, pred_flat, sample_axis=0, event_axis=None)
    total_pcc.append(pcc.numpy())

# è¨ˆç®—å¹³å‡å€¼
avg_rmse = np.mean(total_rmse)
avg_ssim = np.mean(total_ssim)
avg_pcc = np.mean(total_pcc)

print("\n" + "="*60)
print("ğŸ“Š æ”¹é€²å¾Œçš„è©•ä¼°å ±å‘Š")
print("="*60)
print(f"ğŸ”¹ RMSE (å‡æ–¹æ ¹èª¤å·®):     {avg_rmse:.4f}  (è¶Šä½è¶Šå¥½)")
print(f"ğŸ”¹ SSIM (çµæ§‹ç›¸ä¼¼æ€§):     {avg_ssim:.4f}  (è¶Šé«˜è¶Šå¥½)")
print(f"ğŸ”¹ PCC (Pearson ç›¸é—œ):    {avg_pcc:.4f}  (è¶Šé«˜è¶Šå¥½)")
print("="*60)

# æ¯”è¼ƒæ”¹é€²æ•ˆæœ
print("\nğŸ“ˆ èˆ‡åŸå§‹çµæœæ¯”è¼ƒï¼š")
print(f"   RMSE: 0.4457 â†’ {avg_rmse:.4f} (æ”¹å–„: {(0.4457-avg_rmse)/0.4457*100:.1f}%)")
print(f"   SSIM: 0.4420 â†’ {avg_ssim:.4f} (æ”¹å–„: {(avg_ssim-0.4420)/0.4420*100:.1f}%)")
````
